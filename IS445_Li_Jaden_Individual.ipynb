```python
df = pd.read_csv(r"D:\Downloads\archive\abalone1.data.txt", encoding='iso-8859-1', header=None,
                 names=['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 
                        'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings'])
df
```

```python
# Chart 1 Histogram of Abalone Age (Rings)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Rings', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Abalone Age (Rings)')
plt.xlabel('Number of Rings')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Chart 2 Scatter Plot of Abalone Length vs. Diameter
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Length', y='Diameter', color='green')
plt.title('Abalone Length vs. Diameter')
plt.xlabel('Length')
plt.ylabel('Diameter')
plt.grid(True)
plt.show()
```

```python
# Create strip plot
plt.figure(figsize=(8, 6))
sns.stripplot(data=df, x='Sex', y='Rings', jitter=True, palette=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Sex')
plt.ylabel('Number of Rings')
plt.title('Distribution of Abalone Age (Rings) by Sex')
plt.show()
```
