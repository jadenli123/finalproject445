```python
import pandas as pd
import bqplot
import json
import pandas as pd 
import bqplot.pyplot as plt
import re
import numpy as np
import random
from functools import reduce
import seaborn as sns
from ast import literal_eval
# import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Downloads\archive\abalone1.data.txt", encoding='iso-8859-1', header=None,
                 names=['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 
                        'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings'])
df
```

```python
# Take a sample of the first 50 rows
df_sample = df.head(50)

# Chart 1: Scatter Plot of Abalone Length vs. Diameter
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sample, x='Length', y='Diameter', color='green')
plt.title('Abalone Length vs. Diameter (Sample)')
plt.xlabel('Length')
plt.ylabel('Diameter')
plt.grid(True)
plt.show()
```

```python
# Chart 2: Distribution of Abalone by Sex (Sample)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sample, x='Diameter', y='Shell_weight', hue='Sex', palette=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Scatter Plot of Diameter vs. Shell Weight (Sample)')
plt.xlabel('Diameter')
plt.ylabel('Shell Weight')
plt.grid(True)
plt.show()
```

```python
# Convert columns to numeric, replacing non-numeric values with NaN
df_sample['Diameter'] = pd.to_numeric(df_sample['Diameter'], errors='coerce')
df_sample['Shell_weight'] = pd.to_numeric(df_sample['Shell_weight'], errors='coerce')


# Chart 3 Plotting the boxplot for Diameter and Shell_weight with hue as Sex
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_sample, x='Diameter', y='Shell_weight', hue='Sex', palette=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Diameter')
plt.ylabel('Shell Weight')
plt.title('Box Plot of Diameter vs Shell Weight by Sex (Sample)')
plt.show()

```
