```python
import pandas as pd
import bqplot
import json
import pandas as pd 
import bqplot.pyplot as plt
import re
import numpy as np
import random
from functools import reduce
import seaborn as sns
from ast import literal_eval
# import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Downloads\archive\abalone1.data.txt", encoding='iso-8859-1', header=None,
                 names=['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 
                        'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings'])
df
```

```python
#1 Bar Chart of Abalone Sex Distribution:
    
# Calculate sex counts
sex_counts = df['Sex'].value_counts()

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(sex_counts.index, sex_counts.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Distribution of Abalone by Sex')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sex_counts.values, labels=sex_counts.index, colors=['skyblue', 'lightcoral', 'lightgreen'], autopct='%1.1f%%')
plt.title('Distribution of Abalone by Sex')
plt.show()
```

```python
# Create swarm plot
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df, x='Sex', y='Rings', palette=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Sex')
plt.ylabel('Number of Rings')
plt.title('Distribution of Abalone Age (Rings) by Sex')
plt.show()
```
